Then, some file manipulation to get the list of files into a workable form. A lambda function was necessary here to sort the list of files.


```python
List_of_images_to_iter = []
for folder in os.listdir("/content/sample_data"):
  if len(folder) == 1:
    full_path = "/content/sample_data/" + folder
    images = glob.glob(full_path + '/*.png')
    List_of_images_to_iter.append(images)
```

```python
Sorted_list_of_images = []
for List in List_of_images_to_iter:
  sorted_list_two = sorted(List, key = lambda x: int(x[-6:-4]))
  Sorted_list_of_images.append(sorted_list_two)

print(Sorted_list_of_images)
```

```python
List_of_texts_to_iter = []
for folder in os.listdir("/content/sample_data"):
  if len(folder) == 1:
    full_path_two = "/content/sample_data/" + folder
    texts  = glob.glob(full_path_two + '/*.txt')
    List_of_texts_to_iter.append(texts)
```

```python
Sorted_list_of_texts = []
for List in List_of_texts_to_iter:
  sorted_list = sorted(List, key = lambda x: int(x[-10:-8]))
  Sorted_list_of_texts.append(sorted_list)
    
print(Sorted_list_of_texts)
```
