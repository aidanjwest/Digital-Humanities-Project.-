```python
L_1971 = []  #Lists of documents. Grouped by year. 
L_1972 = []
L_1973 = []
L_1974 = []
```

**Grouping the files by year.**

```python
file_list = glob.glob(os.path.join(os.getcwd(), '/content/' , "*.txt"))

for file_path in file_list:
  with open(file_path) as file_input:
    if file_path[9:13] == "1971":
      L_1971.append(file_input.read())
    elif file_path[9:13] == "1972":
      L_1972.append(file_input.read())
    elif file_path[9:13] == "1973":
      L_1973.append(file_input.read())
    elif file_path[9:13] == "1974":
      L_1974.append(file_input.read())
```

** Clean-up function to tokenize, remove stopwords, stem and remove non-words from the text files.**

```python
def clean_up(Output_list, input_list):
  for text in input_list:
    words = nltk.word_tokenize(text)
    words_2 = [word.lower() for word in words if word.isalpha()]
    words_3 = [word for word in words_2 if word not in nltk.corpus.stopwords.words('english')]
    words_4 = [hobj.stem(word) for word in words_3]
    words_5 = [word for word in words_4 if word != []]
    words_6 = [str(word[0])[2:-1] for word in words_5]
    words_7 = [word for word in words_6 if len(word) != 1]
    Output_list.append(words_7)
```

```python
All_docs = []
for file_path in file_list:
  with open(file_path) as file_input:
     All_docs.append(file_input.read())
```

```python
Dishy_Rishi = corpora.Dictionary(All_docs_DONE)
corpus_christi = [Dishy_Rishi.doc2bow(text) for text in All_docs_DONE]
```

```python
lsi = models.LsiModel(corpus_christi, id2word=Dishy_Rishi)
```

```python
from gensim import similarities
index = similarities.MatrixSimilarity(lsi[corpus_christi])
```

**Function to obtain the similarity for an inputted document, compared to the rest of the corpus. In our case, the corpus was only 16 documents, so 16 measures were given. Naturally, when compared to itself, the document has a similarity score of ~99%.**

```python
def get_similarity(path):

  with open(path , 'r') as infile:
        doc = infile.read()

  vec_bow = Dishy_Rishi.doc2bow(doc.lower().split())
  vec_lsi = lsi[vec_bow]  # convert the query to LSI space

  sims = index[vec_lsi]  # perform a similarity query against the corpus

  return list(enumerate(sims))
```

**Creating a simple bar chart to visualize the similarity score.**

```python
def create_bar_chart(x):
  x_axis = []
  y_axis = []
  for i, j in get_similarity(x):
    x_axis.append(i)
    y_axis.append(j)

  import matplotlib.pyplot as plt

  plt.bar(x_axis,y_axis)
  plt.title('Similarity Measures')
  plt.xlabel('Document')
  plt.ylabel('Similarity score')
  plt.show()
```
